# About the Libraries

Detecto is a Python package that allows the construction of object detection models with a minimum of 5 lines of code. It is built on top of PyTorch but abstracts away some of the steps to facilitate development. It also includes tools for inference on static images and videos, transfer learning on custom datasets, and model serialization to files, among other features. Detecto utilizes a convolutional deep learning architecture for model training, specifically a pre-trained model called Faster R-CNN ResNet-50-FPN.

OpenCV is the most well-known Python package for computer vision. It enables image processing, application of filters, modifications, and more. It offers a wide range of features; for more information, visit the website: https://opencv.org/

Matplotlib is the most famous Python package for data visualization in charts. While it is commonly used for plotting graphs, it is also useful for certain commands in conjunction with OpenCV.

## Train model

# Explanation of the Code Below
The dataset has two divisions: one division is actually used for the algorithm to learn, and the other serves as validation when the model is tested with data it has never seen before.

The parameters and their chosen values for model training are:

120 epochs, meaning that the model iterated through the training dataset 120 times to update the internal parameters of the network.
Learning Rate, which, translated into Portuguese, means the learning rate. It is a number, or more precisely, a rate that indicates how much a model will change in response to the estimated error every time the model weights are updated. </br>
For other parameters, default values from the library were chosen, as they are more complex parameters. It was decided to leave them as default values since the library was specifically designed for object detection. Therefore, it is believed that default values are the best choices. For more information, please refer to the library documentation (LEO, include the documentation as a reference here; </br> the link is: https://detecto.readthedocs.io/en/latest/api/core.html?highlight=fit#detecto.core.Model.fit).

```python
dataset = Dataset('/content/sugarbeetandweeds/data/train/annotation','/content/sugarbeetandweeds/data/train/image')
val = Dataset('/content/sugarbeetandweeds/data/val/annotation', '/content/sugarbeetandweeds/data/val/image')
loader = DataLoader(dataset=dataset, batch_size = 2)
your_labels = ['sugarbeet', 'weed']
model1 = Model(your_labels)
```

```python
losses = model1.fit(loader, val, epochs=120, learning_rate=0.001, verbose=True)
```

## Evaluation

O Gráfico abaixo mostra como o treinamento se comportou em relação a "Loss", que é um termo que se refere ao erro médio do algoritmo, ao quanto ele está errando para as predições que ele faz. A função de perca serve como um otimizador, dependendo do que o cientista, observador ou seja quem for que está construindo a rede neural necessita, ele precisa saber se a taxa de perca que ele está tendo é boa ou não. Para avaliação do modelo, este foi a única métrica utilizada para avaliar a performance do modelo.

```python
plt.plot(losses)
plt.show()
```
